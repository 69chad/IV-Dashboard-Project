```python
!pip install yfinance --upgrade --quiet
import yfinance as yf
import pandas as pd

# SPX se trguje kroz ETF - 'SPY' na Yahoo Finance (za NDX koristi 'QQQ')
ticker = 'SPY'

# Dohvati objekt
opt = yf.Ticker(ticker)

# Uzmi prvu expiraciju
exp = opt.options[0]  # npr. '2025-08-09'

# Dohvati opcije za taj datum
chain = opt.option_chain(exp)
calls = chain.calls
puts = chain.puts

# Dohvati spot cijenu
spot_price = opt.history(period='1d')['Close'][-1]

print(f"Spot price for {ticker}: {spot_price:.2f}")
```

```python
# Filtriraj call i put opcije blizu 25-delta strikeova
# 25-delta put je otprilike 10-15% ispod spota, call 10-15% iznad

put_strike = spot_price * 0.85
call_strike = spot_price * 1.15

# Najbliži strike za put
put_row = puts.iloc[(puts['strike'] - put_strike).abs().argsort()[:1]]
call_row = calls.iloc[(calls['strike'] - call_strike).abs().argsort()[:1]]

# Izračunaj skew (koristimo 'impliedVolatility' kao IV)
put_iv = put_row['impliedVolatility'].values[0] * 100
call_iv = call_row['impliedVolatility'].values[0] * 100
skew = put_iv / call_iv

# Prikaz
df = pd.DataFrame({
    'type': ['25D Put', '25D Call'],
    'strike': [put_row['strike'].values[0], call_row['strike'].values[0]],
    'IV (%)': [round(put_iv, 2), round(call_iv, 2)]
})

display(df)
print(f"Skew (put IV / call IV): {skew:.3f}")

if skew < 1.05:
    print("⚠️ Skew < 1σ → Potential tail risk setup.")
```

```python
def get_iv_skew(opt, exp, spot):
    chain = opt.option_chain(exp)
    puts = chain.puts
    calls = chain.calls

    put_strike = spot * 0.85
    call_strike = spot * 1.15

    put_row = puts.iloc[(puts['strike'] - put_strike).abs().argsort()[:1]]
    call_row = calls.iloc[(calls['strike'] - call_strike).abs().argsort()[:1]]

    put_iv = put_row['impliedVolatility'].values[0] * 100
    call_iv = call_row['impliedVolatility'].values[0] * 100
    skew = put_iv / call_iv

    return round(put_iv, 2), round(call_iv, 2), round(skew, 3)
```

```python
iv_history = []

while True:
    clear_output(wait=True)
    put_iv, call_iv, skew = get_iv_skew(opt, exp, spot_price)
    timestamp = pd.Timestamp.now()

    iv_history.append({'time': timestamp, 'put_iv': put_iv, 'call_iv': call_iv, 'skew': skew})
    df = pd.DataFrame(iv_history)

    print(f"⏱ {timestamp.strftime('%H:%M:%S')} | SPY Spot: {spot_price:.2f}")
    print(f"Put IV (25D approx): {put_iv:.2f}%")
    print(f"Call IV (25D approx): {call_iv:.2f}%")
    print(f"Skew (Put/Call IV): {skew:.3f}")

    if skew < 1.05:
        print("⚠️ Skew < 1σ → Tail risk environment!")

    # Plot
    df.set_index('time')[['put_iv', 'call_iv']].plot(figsize=(10,4), title='25D IV Live')
    plt.grid(True)
    plt.ylim(10, 40)
    plt.show()

    df.set_index('time')['skew'].plot(figsize=(10,2), color='orange', title='Skew Over Time')
    plt.grid(True)
    plt.axhline(1.05, linestyle='--', color='red', alpha=0.5)
    plt.ylim(0.8, 1.5)
    plt.show()

    time.sleep(30)  # refresh every 30 seconds
```

```python
import yfinance as yf
import pandas as pd
import matplotlib.pyplot as plt

# Odaberi ticker (SPY proxy za SPX)
ticker = 'SPY'
opt = yf.Ticker(ticker)

# Uzmi najbližu expiraciju
exp = opt.options[0]
chain = opt.option_chain(exp)

# Spot cijena
spot_price = opt.history(period='1d')['Close'][-1]
```

```python
# Kombiniraj puts i calls
puts = chain.puts[['strike', 'openInterest']]
calls = chain.calls[['strike', 'openInterest']]

puts = puts.rename(columns={'openInterest': 'put_OI'})
calls = calls.rename(columns={'openInterest': 'call_OI'})

# Merge
oi = pd.merge(puts, calls, on='strike', how='outer').fillna(0)
oi['total_OI'] = oi['put_OI'] + oi['call_OI']
oi = oi.sort_values('strike')

# Najveći OI strikeovi (top 10)
top_oi = oi.sort_values('total_OI', ascending=False).head(10)
display(top_oi)
```

```python
plt.figure(figsize=(12,5))
plt.bar(oi['strike'], oi['put_OI'], width=1.0, label='Put OI', alpha=0.5, color='red')
plt.bar(oi['strike'], oi['call_OI'], width=1.0, label='Call OI', alpha=0.5, color='green')
plt.axvline(spot_price, color='blue', linestyle='--', label=f'Spot: {spot_price:.2f}')
plt.title(f'{ticker} Option Open Interest Map (Exp: {exp})')
plt.xlabel('Strike')
plt.ylabel('Open Interest')
plt.legend()
plt.grid(True)
plt.show()
```

```python
for _, row in top_oi.iterrows():
    strike = row['strike']
    if abs(spot_price - strike) < spot_price * 0.01:  # ako je unutar 1%
        print(f"⚠️ Spot is approaching high-OI strike: {strike} → expect possible dealer gamma flip.")
```

```python
def calculate_iv_rank(iv_series):
    iv_min = iv_series.min()
    iv_max = iv_series.max()
    current_iv = iv_series.iloc[-1]
    iv_rank = (current_iv - iv_min) / (iv_max - iv_min) if iv_max != iv_min else 0
    return iv_rank
```

```python
import numpy as np

# Dohvati prve dvije expiracije (kratka i srednja)
exp_dates = opt.options[:2]

# Funkcija za prosjek IV za svaku expiraciju
def get_avg_iv(expiration):
    chain = opt.option_chain(expiration)
    # Kombiniraj put i call IV u jednu seriju
    puts_iv = chain.puts['impliedVolatility']
    calls_iv = chain.calls['impliedVolatility']
    all_iv = pd.concat([puts_iv, calls_iv])
    return all_iv.mean()

short_iv = get_avg_iv(exp_dates[0])
long_iv = get_avg_iv(exp_dates[1])
```

```python
if short_iv < long_iv:
    vol_term_structure = 'Contango'
else:
    vol_term_structure = 'Backwardation'

print(f"Volatility term structure: {vol_term_structure}")
```

```python
# Za primjer, simuliraj povijesni IV serijal od zadnjih 30 dana (koristeći povijesne closeove)
# Ovdje možeš koristiti stvarne podatke ako ih imaš

# Ovdje je primjer s random podacima - zamijeni s pravim povijesnim IV podacima ako možeš dohvatiti
iv_history_sample = pd.Series(np.random.uniform(0.15, 0.35, 30))

iv_rank = calculate_iv_rank(iv_history_sample)
print(f"IV Rank (30d): {iv_rank:.2f}")
```

```python
# Izračunaj dnevni povrat i standardno odstupanje (volatilnost cijene)
hist = opt.history(period='30d')
hist['returns'] = hist['Close'].pct_change()
volatility_price = hist['returns'].std()

print(f"Price volatility (30d std dev): {volatility_price:.4f}")
```

```python
# Definiraj pragove (možeš prilagoditi)
iv_rank_threshold = 0.2
volatility_price_threshold = 0.01  # niska volatilnost cijene

if iv_rank < iv_rank_threshold and volatility_price < volatility_price_threshold and vol_term_structure == 'Contango':
    print("🚀 Signal: Buy straddles (IV low + range compression + contango)")
else:
    print("No trade signal at the moment.")
```

```python
import time

def find_strike(strike_list, target):
    # Pronađi closest strike to target
    return min(strike_list, key=lambda x: abs(x - target))

def simulate_asymmetric_trade(opt, spot_price, hold_days=3):
    exp = opt.options[0]  # koristimo najbližu expiraciju
    chain = opt.option_chain(exp)

    # Strikeovi i OI
    puts = chain.puts[['strike', 'lastPrice']]
    calls = chain.calls[['strike', 'lastPrice']]

    # Ciljani strikeovi
    put_strike_target = spot_price * 0.9
    call_strike_target = spot_price * 1.1

    put_strike = find_strike(puts['strike'], put_strike_target)
    call_strike = find_strike(calls['strike'], call_strike_target)

    # Nabavi cijene opcija na ulazu
    put_price_entry = puts.loc[puts['strike'] == put_strike, 'lastPrice'].values[0]
    call_price_entry = calls.loc[calls['strike'] == call_strike, 'lastPrice'].values[0]

    print(f"Entering trade:")
    print(f"  Buy PUT {put_strike} @ {put_price_entry}")
    print(f"  Buy CALL {call_strike} @ {call_price_entry}")

    # Simuliraj čekanje (drži trade)
    print(f"Holding for {hold_days} days...")
    time.sleep(2)  # za test umjesto pravih dana

    # Osvježi chain na kraju držanja
    # (u pravom slučaju čekaš dane i refresh)
    chain_end = opt.option_chain(exp)
    puts_end = chain_end.puts[['strike', 'lastPrice']]
    calls_end = chain_end.calls[['strike', 'lastPrice']]

    put_price_exit = puts_end.loc[puts_end['strike'] == put_strike, 'lastPrice'].values[0]
    call_price_exit = calls_end.loc[calls_end['strike'] == call_strike, 'lastPrice'].values[0]

    print(f"Exiting trade:")
    print(f"  PUT {put_strike} @ {put_price_exit}")
    print(f"  CALL {call_strike} @ {call_price_exit}")

    # Izračun P/L (pretpostavimo kupovina po ulazu, prodaja po izlazu)
    pl_put = put_price_exit - put_price_entry
    pl_call = call_price_exit - call_price_entry
    total_pl = pl_put + pl_call

    print(f"Trade P/L: {total_pl:.2f}")

    return {
        'put_strike': put_strike,
        'call_strike': call_strike,
        'put_price_entry': put_price_entry,
        'call_price_entry': call_price_entry,
        'put_price_exit': put_price_exit,
        'call_price_exit': call_price_exit,
        'total_pl': total_pl
    }

# Pokreni simulaciju
trade_result = simulate_asymmetric_trade(opt, spot_price, hold_days=3)
```
